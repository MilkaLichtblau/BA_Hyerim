```python
import numpy as np
import scipy.stats as stats
import scipy.stats as stats
import matplotlib.pyplot as pl
import pandas as pd
import csv
import json
'''
0: German
1: Other
2: Asylum, refugees
3: EU Country
'''
data = []
a = 1
data_csv =open('mathe_note.csv', 'r')

data_reader = csv.reader(data_csv)

for line in data_reader:
#     print line
    if(line[1] == '1' or line[1] == '2'):
        data.append({'Group':0, 'Math': int(line[0]), 'Index':a})
        a = a+1    
    elif(line[1] == '3'):    
        data.append({'Group':3, 'Math': int(line[0]), 'Index':a})
        a = a+1    
    elif(line[1] == '4'):
        data.append({'Group':2, 'Math': int(line[0]), 'Index':a})
        a = a+1    
    elif(line[1] == '5' or line[1] == '6' or line[1] == '8' or line[1] == '9'):    
        data.append({'Group':1, 'Math': int(line[0]), 'Index':a})
        a = a+1    
for j in data:
    if (j['Math'] == 1):
        j['Math'] = 5
    elif (j['Math'] == 2):
        j['Math'] = 4
    elif (j['Math'] == 3):
        j['Math'] = 3
    elif (j['Math'] == 4):
        j['Math'] = 2
    elif (j['Math'] == 5):
        j['Math'] = 1    
        
data_csv.close()  
pd.DataFrame(data).to_json("math_note.json", orient='records', lines=True)
```

```python
import numpy as np
import scipy.stats as stats
import scipy.stats as stats
import matplotlib.pyplot as pl
import pandas as pd
import csv
import json

i = 0

data = []

data_csv =open('GermanCredit_sex.csv', 'r')

data_reader = csv.reader(data_csv)


status = np.loadtxt('german.data-numeric.txt')
status_sex = [item[6] for item in status]

# DurationMonth   CreditAmount   score   sex

'''
0: Male single
1: Female: divorced/separated/married
2: Male divorced/separated
3: Male married/widowed
'''

for line in data_reader:
#     print line
    if(status_sex[i]==4.0):
        data.append({'Group': 3, 'Score':float(line[2])})
    elif(status_sex[i]==1.0):
        data.append({'Group': 2, 'Score':float(line[2])})
    elif(status_sex[i]==3.0):
        data.append({'Group': 0, 'Score':float(line[2])})
    elif(status_sex[i]==2.0):
        data.append({'Group': 1, 'Score':float(line[2])})   
    i = i+1

data_csv.close()  
data
# len(data)
pd.DataFrame(data).to_json("german_credit_sex_new.json", orient='records', lines=True)
```

```python
import numpy as np
import scipy.stats as stats
import scipy.stats as stats
import matplotlib.pyplot as pl
import pandas as pd
import csv
import json

F=open('german.data.txt',"r")

header = "Existing checking account   Duration in month   Credit history   Purpose   Credit amount   Savings account/bonds   Present employment since   Installment rate   Personal status and sex   Other debtors   Present residence   Property   Age   Other installment   Housing   Existing credits   Job   Number of people being liable   Telephone   Foreign worker   Score"
header = header.replace("   ", ";")

f=open('GermanCredit_age35.csv', "r")

every = []

for i in f:
    every.append(i[:-2].split(','))

scores = np.array(every)[:,2][1:]
j = 0    
with open('csvfile.csv','wb') as file:
    file.write(header)
    file.write("\n")
    for line in F:
        file.write(line[:-1].replace(" ",";"))
        file.write(';')
        file.write(scores[j])
        file.write('\n')
        j = j+1
    file.close()    
```
